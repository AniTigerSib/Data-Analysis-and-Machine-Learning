```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import pandas as pd
```

### Task 1

Для выполнения задания используется набор данных из файла  titanic_train.csv

```python
df = pd.read_csv('../../resources/lab1_3/titanic_train.csv')
```

### Task 2

Построить график зависимости средней стоимости билета от возраста

```python
df[['Age', 'Fare']].groupby('Age').mean().plot()
```

### Task 3

Построить гистограмму распределения пассажиров по классу обслуживания

```python
df['Pclass'].hist(orientation='horizontal', bins=3)
```

### Task 4

Построить круговую диаграмму распределения пассажиров по атрибуту Survived (0/1)

```python
df[['Survived']].groupby(['Survived']).agg(count=('Survived', 'size')).plot(kind='pie', y='count', autopct='%1.1f%%')
```

### Task 5

Построить диаграмму scattered стоимость билета / возраст

```python
df.plot(kind='scatter', x='Age', y='Fare', alpha=0.5)
```

### Task 6

Построить диаграмму Box зависимости стоимости билета от пола пассажира

```python
df.plot(kind='box', column='Fare', by='Sex', vert=False, grid=True)
```

### Task 4

Объединить оба набора данных в один датафрейм (merge)

```python
# unified = male.merge(female, on='Name', how='inner').fillna(0).sort_values(by='Name')
unified = pd.concat([male, female], ignore_index=True).sort_values(by='Name')
unified
```

### Task 5

Выделить последнюю букву в имени. Для каждой буквы подсчитать количество мужских и женских имен, которые на нее оканчиваются.

```python
unified['LastLetter'] = unified['Name'].str[-1]
unified.head()
```

```python
counts = unified.groupby(['LastLetter', 'Sex']).size().unstack(fill_value=0)
counts
```

### Task 6

Написать функцию, которая принимает имя и возвращает обращение в зависимости от пола ("Г-н" или  "Г-жа")

```python
def greet(name: str, sex: str) -> str:
    if sex.lower() == "male":
        return f"Г-н {name}"
    elif sex.lower() == "female":
        return f"Г-жа {name}"
    else:
        return name


print(greet("Иван", "male"))
print(greet("Мария", "female"))
```
